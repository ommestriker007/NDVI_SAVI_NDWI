```python
import geemap.geemap as geemap
import ee

ee.Authenticate()
ee.Initialize(project = 'neon-feat-313910')
```

```python
from ast import Add
base = geemap.Map()
Bangalore = ee.FeatureCollection('projects/neon-feat-313910/assets/Bangalore')
base.addLayer(Bangalore, {}, 'Bangalore')
```

```python
Landsat_5 = ee.ImageCollection("LANDSAT/LT05/C02/T1_TOA").select('B4','B3','B2').filterDate('1990-02-01', '1990-05-15').filterMetadata('CLOUD_COVER', 'less_than', 10)
Landsat_1990 = Landsat_5.mean().clip(Bangalore)
Landsat_1990_NDVI = Landsat_1990.normalizedDifference(['B4','B3']).rename('NDVI')
# landsat_1990_SAVI = Landsat_1990.expression('(B4 - B3) * 1.5 / (B4 + B3 + 0.5)').rename('SAVI')
landsat_1990_SAVI = Landsat_1990.expression(
    '(NIR - RED) * (1 + L) / (NIR + RED + L)',
    {'NIR': Landsat_1990.select('B4'),  # NIR band
     'RED': Landsat_1990.select('B3'),  # Red band
     'L': 0.5}).rename('SAVI')
landsat_1990_NDWI = Landsat_1990.normalizedDifference(['B4','B2']).rename('NDWI')
base.addLayer(Landsat_1990_NDVI, {'min': -0.1, 'max': 0.3, 'palette': ['navy', 'brown','red','orange','yellow','green']}, '1990_NDVI')
base.addLayer(landsat_1990_SAVI, {'min': -1, 'max': 1, 'palette': ['navy', 'brown','red','orange','yellow','green']}, '1990_SAVI')
base.addLayer(landsat_1990_NDWI, {'min': -1, 'max': 1, 'palette': ['navy', 'brown','red','orange','yellow','green']}, '1990_NDWI')
#'#640000', '#ff0000', '#ffff00', '#00c800', '#006400''green','red','yellow'
```

```python
landsat_7 = ee.ImageCollection(ee.ImageCollection("LANDSAT/LE07/C02/T1_TOA")).select('B4','B3','B2').filterDate('2000-02-01', '2000-05-15').filterMetadata('CLOUD_COVER', 'less_than', 10)
Landsat_2000 = landsat_7.mean().clip(Bangalore)
Landsat_2000_NDVI = Landsat_2000.normalizedDifference(['B4','B3']).rename('NDVI')
# landsat_2000_SAVI = Landsat_2000.expression('1.5 * (B4 - B3) / (B4 + B3 + 0.5)').rename('SAVI')
landsat_2000_SAVI = Landsat_2000.expression(
    '(NIR - RED) * (1 + L) / (NIR + RED + L)',
    {'NIR': Landsat_2000.select('B4'),  # NIR band
     'RED': Landsat_2000.select('B3'),  # Red band
     'L': 0.5}).rename('SAVI')
landsat_2000_NDWI = Landsat_2000.normalizedDifference(['B4','B2']).rename('NDWI')
base.addLayer(Landsat_2000_NDVI, {'min': -1, 'max': 1, 'palette': ['navy', 'brown','red','orange','yellow','green']}, '2000_NDVI')
base.addLayer(landsat_2000_SAVI, {'min': -1, 'max': 1, 'palette': ['navy', 'brown','red','orange','yellow','green']}, '2000_SAVI')
base.addLayer(landsat_2000_NDWI, {'min': -1, 'max': 1, 'palette': ['navy', 'brown','red','orange','yellow','green']}, '2000_NDWI')
```
